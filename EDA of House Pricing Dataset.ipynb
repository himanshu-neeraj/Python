```python
sns.set_style("whitegrid")
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='furnishingstatus', data=housing_data)
plt.title('Count of Furnishing Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(data=housing_data, x='price', kde=True)
plt.title('Distribution of House Prices')
plt.xlabel('Price')
plt.show()
```

```python
sns.relplot(x='area', y='price', data=housing_data, height=6, aspect=1.2)
plt.title('Relationship between Area and Price')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='bedrooms', hue='furnishingstatus', data=housing_data)
plt.title('Bedroom Count grouped by Furnishing Status')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='stories', data=housing_data)
plt.title('Count of Houses by Number of Stories')
plt.show()
```
