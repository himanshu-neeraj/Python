```python
# To print even numbers between 1 and 50 and skip multiples of 10 and stop if number > 40

for num in range(1, 51):
    if num % 10 == 0:  # Check if number is a multiple of 10
        continue
    if num > 40:  # Check if number is greater than 40
        break
    if num % 2 == 0:  # Check if number is even
        print(num)
```

```python
#for palindrome series 
num=int(input('enter number:'))
rev=0
x=num
while(num>0):
    rev=(rev*10)+num%10
    num=num//10
if(x==rev):
    print("givern number is palindrome")
else:
    print("given number is not a palindrome")
```

```python
#for armstrong series 
n=int(input("enter the number:"))
x=n
sum=0
while(n>0):
    sum=sum+(n%10)*(n%10)*(n%10)
    n=n//10
if x==sum:
    print('number is armstrong')
else:
    print('number is not armstrong')
```

```python
#for fibonacci series 
n=int(input("Enter the limit number"))
x=0
y=1
z=0
while(z<=n):
  print(z)
  x=y
  y=z
  z=x+y
```

```python
# Factorial using Function

def factorial_recursive(n):
    if not isinstance(n, int) or n < 0:
        return "Factorial is not defined for non-integers or negative numbers."
    
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)
```

```python
#Calling the function
factorial_recursive(6)
```
