```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load Data
df = pd.read_csv('Online Sales Data.csv')

# 2. Plot Chart
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# Agar errorbar pe error aaye to 'errorbar=None' ko hata kar 'ci=None' kar dein
sns.barplot(data=df, x='Product Category', y='Total Revenue')

plt.title('Total Revenue by Product Category', fontsize=16)
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))

sns.countplot(data=df, x='Payment Method')
plt.title('Most Used Payment Methods', fontsize=16)
plt.xlabel('Payment Method')
plt.ylabel('Count of Transactions')
plt.show()
```

```python
plt.figure(figsize=(8, 6))

numeric_cols = df[['Units Sold', 'Unit Price', 'Total Revenue']]

sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

plt.figure(figsize=(12, 6))

sns.lineplot(data=df, x='Date', y='Total Revenue', estimator='sum', errorbar=None, color='blue')

plt.title('Daily Sales Trend (Time ke saath Sales)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.scatterplot(
    data=df, 
    x='Unit Price', 
    y='Total Revenue', 
    size='Units Sold',       
    sizes=(50, 200),      
)

plt.title('Price vs Revenue Scatterplot', fontsize=16)
plt.xlabel('Unit Price')
plt.ylabel('Total Revenue')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left') 
plt.grid(True)
plt.tight_layout()
plt.show()
```
