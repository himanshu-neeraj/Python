```python
sns.barplot(df["type"])
plt.title("Types of content on Netflix")
```

```python
df.rename(columns={"release_year":"year_added"})
year_counts = df['release_year'].value_counts().sort_index()

sns.lineplot(x=year_counts.index, y=year_counts.values)
plt.title("Content Added Over Years on Netflix")
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.show()
```

```python
sns.countplot(y='rating', data=df)
plt.title("Content Rating Distribution on Netflix")
plt.xlabel("Count")
plt.ylabel("Rating")
plt.show()
```

```python
sns.boxplot(x='type', y='release_year', data=df)
plt.title("Release Year Distribution by Content Type")
plt.show()
```

```python
sns.histplot(df['release_year'], bins=20)
plt.title("Release Year Distribution")
plt.show()
```

```python
sns.kdeplot(df['release_year'])
plt.title("KDE Plot of Release Year")
plt.show()
```

```python
sns.histplot(df['release_year'], bins=20,kde=True)
plt.title("Release Year Distribution")
plt.show()
```

```python
sns.violinplot(x='type', y='release_year', data=df)
plt.title("Violin Plot: Release Year by Type")
plt.show()
```

```python
sns.pairplot(df[['release_year']])
```

```python
sns.heatmap(df[['release_year']].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()   #hence their is only positive corerelation
```

```python
sns.scatterplot(x='release_year', y=df.index, data=df)
plt.title("Scatter Plot Example")
plt.show()
```

```python
#Relational PLot(relplot)
sns.relplot(x="release_year", y=df.index,  col="type", data=df)
plt.show()
```
